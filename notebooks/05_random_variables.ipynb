```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from scipy.stats import norm, binom, poisson
```

El modelo Bernoulli se aplica a un random variable que puede asumir solo dos valores. Para este modelo tenemos:
- $E(X)=p$
- $V(X)=pq$

Donde p es la probabilidad de éxito y q es la probabilidad de fracaso. $E(X)$ significa la esperanza de la variable aleatoria X y $V(X)$ significa la varianza de la variable aleatoria X

El modelo binomial puede considerarse como la suma de n variables Bernoulli independientemente una de la otra donde se cumpla lo siguiente:

1. Debe de haber un número fijo de pruebas repetidas estadisticamente independientes.
2. Cada prueba debe ser una variable bernoulli, es decir puede resultar en un éxito o un fracazo.
3. Todas las pruebas deben de tener identicas probabilidades de éxito.

Modelo binominal discrto y tiene lo siguiente: 
- $E(x)=n*p$
- $V(x)=n*p*q$

Donde n es el número de ensayo o pruebas, p es la probabilidad de éxito y q es la probabilidad de fracazo

Una moneda es lanzada 10 veces. ¿Cuál es la probabilidad de obtener 8 o más caras?

```python
n = 10
p = 0.5 
# P( X>=8) = P(X=8)+
prop = 1 - binom.cdf(7,n,p) #cdf es la funcion de destribuión acumulada que indica P(X=0) + P(X=1) + .. P(X=7)
prop
```

Se sabe que el 90% de los trabajadores de una ciudad industrial son miembros de un sindicato. En un estudio de la situación laboral de la ciudad un economista escoge al azar una muestra de 15 obreros para entrevistas ¿Cual es la probabilidad de que la muestra contenga 12 o menos miembros del sindicato?

```python
p = 0.9
n = 15
# P(X <= 12) = P(x=0) + P(x=1) + ...+ P(x=12) = cdf(12)
prop = binom.cdf(12,n,p)
prop
```

Muchos experimentos no ocurren como un resultado definido de pruebas sino como puntos de tiempo, espacio o volumne al azar. El modelo Poisson es adecuado en estos casos si:

1. El número de ocurrencias del hecho es independiente de una unidad
2. El valor esperado de la variable es proporcional al tamaño de la unidad específica
3. La probabilidad de más de una ocurrencia del hecho de una unidad específica muy pequeña es despreciable

Para este modelo tenemos:
- $E(X)=\lambda$
- $V(X)=\lambda$

Donde $\lambda$ es igual al promedio de números de ocurrencias por unidad especificada. La distribución Poisson es discreta

En un callcenter ocurren 4 llamadas por minuto ($\lambda$)=4 ¿Cuál es la probabilidad de que no ocurra más de 2 llamadas durante los 2 minutos siguientes?

```python
#La P(x>=4) = P(X=4) + P(X=5) +P(X=6)+.....P(X=inf)
#P(X>=4) = 1 - P(X=3) = 1 - P(X=0) - P(X=2) - P(X=3)
n = 3 #Número de éxitos
plambda=8 #Número de ocurrencias en una unidad especificada (4 llamadas por minuto en dos min es 8)
prop = 1 - poisson.cdf(n,plambda)
prop
```

```python
n =2
plambda = 8
#P(x>=4) es decir P(X=4) + P(X=5) + P(X=6)
prop = 1  - poisson.cdf(n, plambda)
prop
```

Un fabricante de tejido de lanas afirma que el promedio de defectos en su producto es 1 por 2 yardas cuadradas. Se extrajo una yarda cuadrada al azar y tiene 3 defectos
¿Cuál es la probabilidad de obtener 3 o más defectos en cualquier yarda cuadrada si la afirmación del fabricante fuera válida?

```python
plambda=0.5
n=3
#P(X>=3) = P(X=3) + P(X=4) + P(X=5) ... P(X=inf)
#P(X>=3) = 1 - P(X<=2) =  - P(X=0) - P(x=1) - P(X=2)
prop = 1 - poisson.cdf(n, plambda)
prop
```

El modelo de probabilidad más frecuentemente usado en el análisis económico y comercial es la distirbución normal, que pueda expresarse en la forma general y estandarizadas. Se dice que una variable aleatoria X tiene una distirbución normal si es continua, si esta en el rango del menos infinito al infinito y si puede expresarse por su media $\mu$ y su desviación estandar $\sigma$

```python
x = np.linspace(-5,5,1000)
pdf = norm.pdf(x)
plt.plot(x,pdf,c="tomato")
plt.title("PDF Distribución Normal")
plt.xlabel("value")
plt.ylabel("density")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-5,5,1000)
cdf = norm.cdf(x)
plt.plot(x,cdf,c="tomato")
plt.title("CDF Distribución Normal")
plt.xlabel("value")
plt.ylabel("probability")
plt.grid(True)
plt.show()
```

```python
n = 2
p = 0.5
x = range(n+1)

pmf = binom.pmf(x,n,p)
plt.stem(x, pmf)
plt.grid(True)
plt.show()
```
