```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
def line_function(x:float)->float:
  return x + 2 #Si X=0 entonces Y=2, Si X = 3 entonces Y= 5. La notación matemática es f(x) = x+2
```

```python
x =np.linspace(-5,5,100)  #100 puntitos espaciados entre el -5 y 5
plt.plot(x, line_function(x), label="Line_function")
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.grid(True)
plt.show()
```

```python
def f(x:float)->float:
  return 2*x + 2
```

```python
x =np.linspace(-5,5,100)  #100 puntitos espaciados entre el -5 y 5
plt.plot(x, f(x), label="f(x)")
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def g(x:float)-> float:
  return x*x +1 #Fución cuadratica
```

```python
x =np.linspace(-5,5,100)  #100 puntitos espaciados entre el -5 y 5
plt.plot(x, g(x), label="g(x)")
plt.axhline(y=0, color="g")
plt.axvline(x=0, color="g")
plt.legend()
plt.grid(True)
plt.plot(0,g(0), "ro")
plt.plot(2,g(2), "ro")
plt.show()
```

```python
def k(x:float)->float: #Función cuadrática con termino lineal
  return x*x - 2*x + 4
```

```python
def dk(x:float)->float:
  h=0.000001
  return (k(x+h)- k(x))/h
```

```python
x = np.linspace(-5,5,1000)
plt.plot(x , k(x), label="k(x)")
plt.plot(x, dk(x), label="dk(x)")
plt.plot(0,dk(0), "ro")
plt.plot(2,dk(2), "ro")
plt.plot(1,dk(1), "ko")  #La derivada de la linea da cero, es el mínimo de la función
plt.axhline(y=3, color="k")
plt.axhline(y=0, color="g")
plt.axvline(x=0, color="g")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def squear(x:float) -> float:
  return x**2 #significa potencia x*x
```

```python
plt.plot(x, squear(x))
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid(True)
plt.title("Gráfica de la función cuadrática en el intervalo 0 al 4")
plt.show()
```

```python
x = np.linspace(-10,10,1000)

def f(x:float)-> float:
  return np.sin(x)

plt.plot(x, f(x))
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid(True)
plt.title("Gráfica de la función Seno en el intervalo del -3 al 3")
plt.show()
```

```python
x = np.linspace(-3,3,1000)

def g(x:float)-> float:
  return np.power(2,x)

plt.plot(x, g(x))
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid(True)
plt.title("Gráfica de la función Potencia en el intervalo del -3 al 3")
plt.show()
```

```python
x = np.linspace(1,1000,1000)

def h(x:float)-> float:
  return np.log10(x)

plt.plot(x, h(x))
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid(True)
plt.title("Gráfica de la función Logaritmica en el intervalo del 1 al 1000")
plt.show()
```
