4. El método de Montecarlo nos permite aproximar una probabilidad en base a la repetición continua de un evento 
   ¿Cuál es la probabilidad de obtener un  número par en un lanzamientode dados?

```python
events = [2,4,6]
space = [1,2,3,4,5,6]
trials = list(range(1,1001))
frec = [] #Aca agregamos los resultados
#Vamos a lanzar(trial) el dado y si esta dentro del espacio muestra es un exito
for trial in trials:
    total = 0
    for i in range(trial):
        if random.choice(space) in events: #Coge un elemento al azar del spacio space si ese valor esta debntro de events es un exito
            total = total + 1
    frec.append(total /trial)

```

```python
plt.plot(trials,frec)
plt.xlabel("Trails")
plt.ylabel("Relative Frequence")
plt.axhline(y=0.5, color="tomato")
plt.grid(True)
plt.show()
#En las primeras veces tal vez no es par, pero si lanzo el dado al infinito la probabilidad de que salga par si va ser 0.5
```
